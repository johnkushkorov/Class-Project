```python
import os
import folium

print(folium.__version__)


import numpy as np

data = (np.random.normal(size=(100, 3)) *
        np.array([[1, 1, 1]]) +
        np.array([[40.7128,  -73.006, 1]])).tolist()
from folium.plugins import HeatMap

m = folium.Map([40.7128, -73.0060], tiles='stamentoner', zoom_start=6)

HeatMap(data).add_to(m)

m

    
```
